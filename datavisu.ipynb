```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv("datasets/dataset_train.csv", index_col = "Index")
```

```python
df
```

```python
df.loc[:,"Hogwarts House"]
```

```python
Ravenclaw = df[df.loc[:,"Hogwarts House"] == "Ravenclaw"].select_dtypes(include=['int64','float64'])
```

```python
Gryffindor = df[df.loc[:,"Hogwarts House"] == "Gryffindor"].select_dtypes(include=['int64','float64'])
```

```python
Slytherin = df[df.loc[:,"Hogwarts House"] == "Slytherin"].select_dtypes(include=['int64','float64'])
```

```python
Hufflepuff = df[df.loc[:,"Hogwarts House"] == "Hufflepuff"].select_dtypes(include=['int64','float64'])
```

```python
Hufflepuff
```

```python
plt.title("Arithmancy")
plt.hist(Ravenclaw.loc[:,"Arithmancy"], alpha = 0.5, lw=3, label="Ravenclaw", color="b")
plt.hist(Hufflepuff.loc[:,"Arithmancy"], alpha = 0.5, lw=3, label="Hufflepuff", color="y")
plt.hist(Slytherin.loc[:,"Arithmancy"], alpha = 0.5, lw=3, label="Slytherin", color="g")
plt.hist(Gryffindor.loc[:,"Arithmancy"], alpha = 0.5, lw=3, label="Gryffindor", color="r")
plt.legend()
plt.show()
```

```python
col_names = df.select_dtypes(include=['int64','float64']).columns
```

```python
col_names
```

```python
nb_col_graph = 3
nb_lignes_graph = int(len(col_names)/nb_col_graph) + 1
plt.figure(figsize=(nb_col_graph * 5, nb_lignes_graph * 5))
for i in range(len(col_names)):
    plt.subplot(nb_lignes_graph, nb_col_graph, i + 1)
    plt.title(col_names[i])
    plt.hist(Ravenclaw.loc[:,col_names[i]], alpha = 0.5, lw=3, label="Ravenclaw", color="b")
    plt.hist(Hufflepuff.loc[:,col_names[i]], alpha = 0.5, lw=3, label="Hufflepuff", color="y")
    plt.hist(Slytherin.loc[:,col_names[i]], alpha = 0.5, lw=3, label="Slytherin", color="g")
    plt.hist(Gryffindor.loc[:,col_names[i]], alpha = 0.5, lw=3, label="Gryffindor", color="r")
    plt.legend()
plt.show()
```

```python
col_names
```

```python
nb_col_graph = 3
nb_lignes_graph = int((len(col_names)- 1)/nb_col_graph) + 1
plt.figure(figsize=(nb_col_graph * 7, nb_lignes_graph * 7))
for i in range(len(col_names)):
    plt.subplot(nb_lignes_graph, nb_col_graph, i + 1)
    plt.title("Arithmancy Vs " + col_names[i])
    plt.plot(df.loc[:,"Arithmancy"], df.loc[:,col_names[i]], 'o', label="Dataset")
    plt.xlabel("Arithmancy")
    plt.ylabel(col_names[i])
    plt.legend()
plt.show()
```
